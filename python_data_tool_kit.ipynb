Q2.  How does broadcasting work in NumPy
- Broadcasting is a powerful mechanism in NumPy that allows arithmetic operations to be performed on arrays of different shapes. It enables NumPy to work with arrays of different sizes without the need for explicit replication of data. The rules of broadcasting are as follows:

1. If the arrays have different numbers of dimensions, the shape of the smaller-dimensional array is padded with ones on the left side until both shapes are the same.
2. If the sizes of the dimensions do not match, the array with size 1 in that dimension is stretched to match the size of the other array.
3. If the sizes of the dimensions do not match and neither is 1, an error is raised.

Q6.  What are the differences between NumPy arrays and Python lists
- Data Type Homogeneity:

1. NumPy Arrays: All elements must be of the same data type, which allows for optimized performance and memory usage.
2. Python Lists: Can contain elements of different data types (e.g., integers, strings, objects).
- Performance:

1. NumPy Arrays: Generally faster for numerical computations due to optimized C and Fortran libraries and contiguous memory allocation.
2. Python Lists: Slower for numerical operations because they are more flexible and require more overhead.
- Functionality:

1. NumPy Arrays: Provide a wide range of mathematical functions and operations (e.g., element-wise operations, broadcasting).
2. Python Lists: Do not have built-in support for mathematical operations; you need to use loops or list comprehensions.
- Memory Efficiency:

1. NumPy Arrays: More memory-efficient due to fixed data types and contiguous memory allocation.
2. Python Lists: Less memory-efficient because they store references to objects and can have overhead.

Q8. What Does the Term “Vectorized Operation” Mean in NumPy

1. Performance: They are generally faster than using Python loops because they leverage low-level optimizations.
2. Conciseness: Code is often more readable and concise, as operations can be expressed in a single line.
3. Ease of Use: They simplify the code by eliminating the need for manual iteration and indexing.

Q17.  Explain the difference between apply() and map() in Pandas?
- In Pandas, both apply() and map() are used to apply functions to data, but they have different use cases and behaviors:

1. apply():

- Can be used on both Series and DataFrames.
- Allows for applying a function along a specific axis (rows or columns) in a DataFrame.
- Can accept functions that return a Series or DataFrame, allowing for more complex operations.
- Example: df.apply(np.sum, axis=0) computes the sum of each column in a DataFrame.


2. map():

- Primarily used with Pandas Series.
- Designed for element-wise operations, meaning it applies a function to each individual element in the Series.
- Typically used for transforming or mapping values in a Series (e.g., replacing values or applying a simple function).
- Example: series.map(lambda x: x * 2) doubles each element in the Series.

Q21. Why is NumPy’s array slicing faster than Python’s list slicing?
- NumPy’s array slicing is faster than Python’s list slicing due to several reasons:

1. Contiguous Memory: NumPy arrays are stored in contiguous blocks of memory, which allows for efficient access and manipulation of data. This contrasts with Python lists, which can store elements in non-contiguous memory locations.
2. Optimized C Implementation: NumPy is implemented in C, which allows for lower-level optimizations and faster execution of operations compared to Python's higher-level list operations.
3. View vs. Copy: When slicing a NumPy array, it returns a view of the original array rather than a copy, which saves memory and processing time. In contrast, Python lists create a new list when sliced.
4. Vectorized Operations: NumPy supports vectorized operations, allowing for batch processing of data without the need for explicit loops, further enhancing performance.

```python
# Q4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap

import seaborn as sns
import matplotlib.pyplot as plt

# Create a sample DataFrame
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [5, 6, 7, 8, 9],
    'C': [9, 10, 11, 12, 13]
}
df = pd.DataFrame(data)

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# Q9. Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold?

import pandas as pd
import numpy as np

# Generate a DataFrame
data = {
    'A': np.random.randint(1, 100, 10),
    'B': np.random.rand(10)
}
df = pd.DataFrame(data)

# Filter rows where column 'A' is greater than a threshold (e.g., 50)
threshold = 50
filtered_df = df[df['A'] > threshold]

print("Original DataFrame:")
print(df)
print("\nFiltered DataFrame (A > 50):")
print(filtered_df)
```
